#### Standardize tracking id column

```python
trackaff_df2["tracking_id"] = trackaff_df2["tracking_id"].astype("Int64").astype(str)
..["tracking_id"] = affmatrix_df2["tracking_id"].astype("Int64").astype(str)
```

```python
trackaff_df2["tracking_id"] = trackaff_df2["tracking_id"].str.replace("<NA>", "", regex=False).str.strip()
affmatrix_df2["tracking_id"] = affmatrix_df2["tracking_id"].str.replace("<NA>", "", regex=False).str.strip()
```

### Split tracking_id into date / card / campaign on  trackaff and affmatrix tables

Assume tracking_id as:

first 6 digits = YYMMDD

next 3 digits = card_id

last 3 digits = campaign_id

```python
trackaff_df2["tracking_id_date"] = trackaff_df2["tracking_id"].str[:6]
trackaff_df2["tracking_id_card_id"] = trackaff_df2["tracking_id"].str[6:9]
trackaff_df2["tracking_id_campaign_id"] = trackaff_df2["tracking_id"].str[9:12]

affmatrix_df2["tracking_id_date"] = affmatrix_df2["tracking_id"].str[:6]
affmatrix_df2["tracking_id_card_id"] = affmatrix_df2["tracking_id"].str[6:9]
affmatrix_df2["tracking_id_campaign_id"] = affmatrix_df2["tracking_id"].str[9:12]
```

### Create column to map Currency to GBP

Assumed coversion rates:

GBP = 1.00

EUR = 0.86

USD = 0.79

```python
fx_to_gbp = {
    "GBP": 1.00,
    "EUR": 0.86,
    "USD": 0.79
}
```

```python
trackaff_df2["fx_to_gbp"] = trackaff_df2["currency"].map(fx_to_gbp)
affmatrix_df2["fx_to_gbp"] = affmatrix_df2["currency"].map(fx_to_gbp)
```

```python
trackaff_df2["total_commission_gbp"] = trackaff_df2["total_commission"] * trackaff_df2["fx_to_gbp"]
trackaff_df2["commission_per_ftd_gbp"] = trackaff_df2["commission_per_ftd"] * trackaff_df2["fx_to_gbp"]

affmatrix_df2["commission_earned_gbp"] = affmatrix_df2["commission_earned"] * affmatrix_df2["fx_to_gbp"]
affmatrix_df2["commission_per_ftd_gbp"] = affmatrix_df2["commission_per_ftd"] * affmatrix_df2["fx_to_gbp"]
```
